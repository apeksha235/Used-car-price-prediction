```python
import pandas as pd
from sklearn import preprocessing 
```

```python
df=pd.read_csv(r'C:\Users\Apeksha\Documents\Mini Project\train-data (3).csv')
```

```python

df=df.dropna(how='any',axis=0)
df.isnull().sum()
```

```python
df=pd.get_dummies(df, columns=['Transmission'],drop_first=True,prefix='Transmission')
df=pd.get_dummies(df, columns=['Fuel_Type'],drop_first=True,prefix='Fuel')

dict={
    'First': 1,
    'Second':2,
    'Third':3,
    'Fourth': 4
}
df['Owner_Type']=df.Owner_Type.map(dict)
```

```python
df.head()
```

```python
col='Engine'
df[col]=df[col].str.replace('CC','')
col1='Mileage'
df[col1]=df[col1].str.replace('kmpl','')
df[col1]=df[col1].str.replace('km/kg','')
col2='Power'
df[col2]=df[col2].str.replace('bhp','')
```

```python
df['Mileage'] = df['Mileage'].astype(float)
df['Engine'] = df['Engine'].astype(int)
df['Power'] = df['Power'].astype(float)
df.head()
```

```python
df.to_csv(r'C:\Users\Apeksha\Documents\Mini Project\Used_cars_data.csv',index=False)
```

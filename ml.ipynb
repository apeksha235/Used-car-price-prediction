```python
df = pd.read_csv(
    r"D:\Study\VScode\KJ_SEM4\MP\Used-car-price-prediction\Used_cars_data.csv")
df.head()
```

```python
X = df[['Year','Kilometers_Driven','Mileage','Engine','Power','Seats','Transmission_Manual','Fuel_Diesel','Fuel_LPG','Fuel_Petrol']]
Y = df['Price']
x_train, x_test,y_train,y_test = train_test_split(X,Y,test_size =0.2,random_state=42)
def reg_func(model):
       model.fit(x_train,y_train)
       y_pred=model.predict(x_test)
       score=mean_squared_error(y_test,y_pred)
       print('Mean Squared Error: '+str(score))
import numpy as np

def clean_dataset(df):
    assert isinstance(df, pd.DataFrame)
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.int32)
clean_dataset(X)
```

```python
model= RandomForestRegressor(n_estimators = 200, random_state = 0)
reg_func(model)
```
